## **Purchase Patterns and Demographics: An Online Shopping Case Study**

Description: This dataset contains information on 10,000 customers who made a purchase online, including their personal details, browser information, payment details, and purchase price. The columns include:

- Address (customer address)
- Lot (customer location?)
- AM/PM (time of purchase?)
- Browser Info (browser type and version)
- Company (customer's company name)
- Credit Card (credit card number)
- CC Exp Date (credit card expiration date)
- CC Security Code (credit card security code)
- CC Provider (credit card provider)
- Email (customer email)
- Job (customer job title)
- IP Address (customer IP address)
- Language (customer language preference)
- Purchase Price (price of the purchased item)

This dataset can be used to analyze customer behavior, purchase patterns, and demographic characteristics. It can also be used to study the effectiveness of marketing campaigns, customer segmentation, and personalization strategies.

15. What are the top 5 most popular email providers (e.g. gmail.com, yahoo.com, etc...)

```python
# Extract email provider from the 'Email' column
df['Email Provider'] = df['Email'].str.split('@').str[1]

# Count the occurrences of each email provider
email_provider_counts = df['Email Provider'].value_counts()

# Get the top 5 most popular email providers
top_5_email_providers = email_provider_counts.head(5)

top_5_email_providers
```

```python
top_5_email_providers.plot(kind='bar')
plt.xlabel('Email Provider')
plt.ylabel('Count')
plt.title('Top 5 Most Popular Email Providers')
plt.show()
```
